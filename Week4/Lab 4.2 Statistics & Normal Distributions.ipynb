#### Select the rows where state is New York

```python
# your answer here
temp_ny = temp[temp['State'] == 'New York']
print(temp_ny)
```

#### What is the average of the temperature of cities in New York?

```python
# your answer here
avg_NY = round(np.mean(temp_ny['Temperature']),2)
print(avg_NY)
```

#### We want to know cities and states with Temperature above 15 degress Celcius

```python
# your answer here
print(' '.join(temp[temp['Temperature']>15]['City'].unique()))
print(' '.join(temp[temp['Temperature']>15]['State'].unique()))
```

#### Now, return only the cities that have a temperature above 15 degress Celcius

```python
# your answer here
print(' '.join(temp[temp['Temperature']>15]['City'].unique()))
```

#### We want to know which cities have a temperature above 15 degrees Celcius and below 20 degrees Celcius

*Hint: First write the condition then select the rows.*

```python
# your answer here
print(' '.join(temp[(temp['Temperature']>15) & (temp['Temperature']<20)]['City']))
```

#### Find the mean and the standard deviation of the temperature of each state.

*Hint: Use functions from Data Manipulation lesson*

```python
# your answer here
avg_CT = round(np.mean(temp[temp['State']=='Connecticut']['Temperature']),2)
avg_NJ = round(np.mean(temp[temp['State']=='New Jersey']['Temperature']),2)

std_NY = round(np.std(temp_ny['Temperature']),2)
std_CT = round(np.std(temp[temp['State']=='Connecticut']['Temperature']),2)
std_NJ = round(np.std(temp[temp['State']=='New Jersey']['Temperature']),2)

print('New York: ', avg_NY, ', New York standard deviation:', std_NY)
print('Connecticut: ', avg_CT, ', Connecticut standard deviation:', std_CT)
print('New Jersey: ', avg_NJ, ', New Jersey standard deviation:', std_NJ)
```
